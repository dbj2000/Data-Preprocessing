## 고객 특성에 따른 군집화

```python
df = pd.read_csv("Telco_customer_info.csv")
df.head()
```

```python
# customerID는 의미없으므로 인덱스로 설정
# set_index : 컬럼을 인덱스화
df.set_index('customerID', inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, drop_first = True) # 처음 값을 가지는지로 설정
df.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
clusters = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward').fit(df)
```

```python
df['군집정보'] = clusters.labels_
```

```python
df['군집정보'].head()
```

```python
# 군집 그룹에 따른 정보 확인
df.groupby(['군집정보'])[['MonthlyCharges', 'TotalCharges']].mean()
```

```python
df.groupby(['군집정보'])[['SeniorCitizen', 'StreamingTV_Yes']].mean()
```

## 구매기록 기준 고객 군집화

```python
df = pd.read_csv("베스트셀러_도서구매기록.txt", sep="\t", engine="python")
```

```python
df.head()
```

* 회원번호의 중복이 발생할 수 있으므로 이 경우에는 get_dummies는 적합하지 않음
* crosstab사용

```python
matrix_df = pd.crosstab(index = df['회원번호'], columns = df['책제목'])
matrix_df.head()
```

```python
# 군집화 모델 인스턴스화 및 학습
from sklearn.cluster import AgglomerativeClustering as AC
# jaccard는 1이상 값을 1로 취급
# jaccard사용하므로 linkage는 무조건 average
clustering_model = AC(n_clusters=10, affinity="jaccard",linkage="average") 
clustering_model.fit(matrix_df)
```

```python
cluster_labels = clustering_model.labels_
cluster_labels
```

```python
# 회원별 소속 군집 확인
cluster_info = pd.DataFrame({"회원ID":matrix_df.index,
                             "소속군집":cluster_labels})
cluster_info.head()
```

```python
cluster_info['소속군집'].value_counts()
```

* jaccard를 썼기 때문에 ward를 쓰지못해 하나의 군집에 몰리게 됨

```python
matrix_df_with_cluster_info = pd.merge(matrix_df, cluster_info, 
                                       left_index=True,
                                       right_on = '회원ID')
matrix_df_with_cluster_info.head()
```

```python
matrix_df_with_cluster_info.groupby(['소속군집'])[matrix_df.columns].mean()
```

```python
# 군집별로 가장 많이 구매하는 책 확인
matrix_df_with_cluster_info.groupby(['소속군집'])[matrix_df.columns].mean().idxmax(axis=1)
```

# K-평균 군집화 실습

```python
import pandas as pd
import os
os.chdir(r"D:\패스트캠퍼스\데이터전처리\part-2.-탐색적-데이터-분석\2. 탐색적 데이터 분석\데이터")
```

```python
# 데이터 전처리는 앞의 과정과 같음
df = pd.read_csv("Telco_customer_info.csv")
# customerID는 의미없으므로 인덱스로 설정
# set_index : 컬럼을 인덱스화
df.set_index('customerID', inplace=True)
df = pd.get_dummies(df, drop_first = True) # 처음 값을 가지는지로 설정
```

```python
from sklearn.cluster import KMeans
clusters = KMeans(n_clusters = 3).fit(df)
```

```python
# 군집 중심 확인 및 이름 붙이기
pd.DataFrame(clusters.cluster_centers_,
             columns = df.columns,
             index = range(3))
```
